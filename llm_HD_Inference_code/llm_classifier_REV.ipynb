#ILDC Dataset

##LLM Inference on Test Data (after running llm_Inference_ILDC.py on test data)

```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import re

y_true = []
y_pred = []
valid_preds = {'accept','reject'}
with open("/content/mistral-7b-instruct-v0.3_ILDC_start1-0_end-2500_2_28.json", "r", encoding="utf-8") as f:
    data = json.load(f)

for item in data:
    # ground truth
    gt = item["answers"].strip().lower()

    # model response
    resp = item["str_response"] #.strip().split(',')[0].split('.')[0].lower()
    match_pred = re.search(r'[A-Za-z-]+', resp)
    pred = match_pred.group(0).lower() if match_pred else ""

    # remove "not sure"
    # if resp == "not":
    #     continue
    # if pred not in valid_preds:
    #     continue
    y_true.append(gt)
    y_pred.append(pred)

print("Ground truth:", y_true)
print("Predictions:", y_pred)
```

##LLM+CD Inference on Test Data(Refrain Setting)[after running llm_classifier_ILDC.py on test data)




```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import re

y_true = []
y_pred1 = []
valid_preds = {'accept','reject'}

with open("/content/gated_results_mistral_fc_last3_0322_1415.json", "r", encoding="utf-8") as f:
    data = json.load(f)

for item in data:
    # ground truth
    gt = item["answer"].strip().lower()

    # model response
    resp = item["response"]#.strip().split(',')[0].split('.')[0].lower()
    match_pred = re.search(r'[A-Za-z-]+', resp)
    pred = match_pred.group(0).lower() if match_pred else ""

    # if pred not in valid_preds:
    #     continue

    # # # # remove "not sure"
    # if pred == "not":
    #     continue

    y_true.append(gt)
    y_pred1.append(pred)
print("Ground truth:", y_true)
print("Predictions:", y_pred1)
```

##LLM+CD Inference on Test Data(Reverse Setting)

```python
for i in range(len(y_pred1)):
  if y_pred1[i] == "not":
      if y_pred[i] == "accept":
          y_pred1[i] = "reject"
      elif y_pred[i] == "reject":
          y_pred1[i] = "accept"
```
